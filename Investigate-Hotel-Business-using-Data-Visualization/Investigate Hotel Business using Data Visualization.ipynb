```python
#Libraries

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read dataset
url = "https://raw.githubusercontent.com/denindrap23/Machine-Learning/refs/heads/main/Investigate-Hotel-Business-using-Data-Visualization/hotel_bookings_data.csv"

df = pd.read_csv(url)
print(df.shape)
df.head()
```

```python
#info dataframe
df.info()
```

```python
#deskripsi statistik
df.describe()
```

# 1. Data Preprocessing

## 1.1 Mengatasi data null

```python
#cek jumlah data null
df.isnull().sum()
```

Terdapat 4 kolom yang memiliki data null:
- `children`: 4 data
- `city`: 488 data
- `agent`: 16340 data
- `company`: 112593 data

```python
#Mengisi data null

#Kolom children dapat diisi dengan angka 0 menggambarkan tidak adanya anak dalam pemesanan hotel
df['children'] = df['children'].fillna(0)

#Kolom city dapat diisi dengan "unknown" karena tidak diketahui negara asal dari pemesan hotel
df['city'] = df['city'].fillna('unknown')

#Kolom agent dan company dapat diisi dengan angka 0 sebagai "undefined" value, karena pada agent dan company tidak terdapat angka 0 (min: 1, max:6)
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)

df.isnull().sum()
```

## 1.2 Mengganti value yang tidak sesuai

```python
#data "meal" mengandung value "Undefined", yang sebenarnya adalah "No Meal"
df.replace({'meal': {"Undefined": "No Meal"}}, inplace=True)
```

## 1.3 Menghapus data yang tidak diperlukan

```python
# Membuat data total guest yaitu jumlah dari adult, children dan babies
df['guest'] = df['adults'] + df['children'] + df['babies']
print("Data with 0 guest: {0} out of {1} all data".format(df[df['guest']==0].shape[0], df.shape[0]))

# Kita akan hapus data yang memiliki 0 guest
df_preprocessing = df[df["guest"] > 0].copy()
df_preprocessing.shape
```

## 1.4 Hasil data preparation

```python
#Membuat data preparation
df_prep = df[df['guest']>0].copy()

print("before pre-processing: ", df.shape[0])
print("after pre-processing: ", df_prep.shape[0])
print("jumlah data yang dibuang: ", df.shape[0]-df_prep.shape[0])
```

# 2. Analisis Jumlah Pemesanan Hotel per Bulan berdasarkan tipe Hotel

```python
# Melakukan copy pada data hasil pre-processing untuk visualisasi
df1 = df_prep.copy()

# Membuat data agregasi untuk menghitung jumlah pemesanan hotel per bulan berdasarkan tipe hotel
df1_agg = df1.groupby(['hotel', 'arrival_date_month'])['arrival_date_year'].agg(['nunique', 'count']).reset_index()
df1_agg.columns = ['hotel_type', 'arrival_month', 'unique_year', 'num_booking'] # mengubah nama kolom tabel
df1_agg
```

Perhatikan data bulan Juli dan Agustus diambil dari data 3 tahun lalu, sedangkan bulan-bulan lain hanya diambil dari data 2 tahun lalu, perlu dilakukan normalisasi agar perbandingannya sesuai

```python
# normalisasi untuk mendapatkan jumlah rata-rata pemesanan di suatu bulan pada tahun yang berbeda
df1_agg['average_num_booking'] = round(df1_agg['num_booking']/df1_agg['unique_year'])
df1_agg.head()
```

```python
# Mengurutkan data berdasarkan bulan, serta membuat singkatan Bulan agar mudah divisualisasikan
ordered_months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
                  "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

df1_agg['month'] = df1_agg['arrival_month'].str[:3]
df1_agg['month'] = pd.Categorical(df1_agg['month'], categories=ordered_months, ordered=True)

df1_agg.head()
```

```python
df1_agg['average_num_booking'].max()
```

Tujuan dari plot yang ingin ditampilkan adalah untuk membandingkan 2 tipe hotel dalam setiap bulannya. Yakni tipe hotel Kota dan tipe hotel Resort dari bulan Januari sampai Desember. Oleh karena itu, jenis plot yang tepat adalah line plot.

```python
# plot awal menggunakan lineplot dari library seaborn
plt.figure(figsize=(15,10))
sns.lineplot(x='month', y='average_num_booking', hue='hotel_type',
             size="hotel_type", sizes=(2.5,2.5), data=df1_agg)

plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid()
plt.legend(title='Hotel Type', title_fontsize=15, prop={'size':13})

plt.xlabel('Arrival Month', fontsize=15)
plt.ylabel('Average Number of Booking', fontsize=15)
plt.ylim(0,df1_agg['average_num_booking'].max()+500)

plt.axvline(4, ls='--', color='red')
plt.axvline(6, ls='--', color='red')
plt.text(x=4.5, y=df1_agg['average_num_booking'].max()+100, s='Holiday \nSeason', fontsize=16, color='red')
plt.stackplot(np.arange(4,7,1), [[df1_agg['average_num_booking'].max()+500]], color='indianred', alpha=0.3)

plt.axvline(9, ls='--', color='red')
plt.axvline(11, ls='--', color='red')
plt.text(x=9.5, y=df1_agg['average_num_booking'].max()+100, s='Holiday \nSeason', fontsize=16, color='red')
plt.stackplot(np.arange(9,12,1), [[df1_agg['average_num_booking'].max()+500]], color='indianred', alpha=0.3)

plt.text(x=-0.5, y=df1_agg['average_num_booking'].max()+900, s="Both Hotels Have More Guests During Holiday Season", 
         fontsize=20, fontweight='bold')
plt.text(x=-0.5, y=df1_agg['average_num_booking'].max()+550, s="City Hotel has the decreased guests in August and September, while both hotels have less \ncustomer during not holiday season (Jan-Mar)", 
         fontsize=18)

plt.tight_layout()
plt.savefig('investigate_hotel_business_v01.png', dpi=200)
```

# 3. Analisa Pengaruh Durasi Menginap terhadap tingkat Pembatalan Pemesanan Hotel

```python
# Melakukan copy pada data hasil pre-processing untuk visualisasi
df2 = df_prep.copy()

# Melakukan aggregasi data untuk mendapatkan durasi menginap berdasarkan jenis hotel
df2['total_nights'] = df2['stays_in_weekend_nights'] + df2['stays_in_weekdays_nights'] #menghitung total malam yang dipesan

sns.histplot(data=df2, x="total_nights", binwidth=5, color="pink")
```

Berdasarkan pada histogram di atas, dapat dilihat bahwa persebaran data terlihat right-skewed. Oleh karena itu dilakukan pengelompokkan untuk membuat persebarannya menjadi lebih merata. Pengelompokkan tersebut dilakukan untuk total jumlah malam yang mencapai 3 minggu maka dikelompokkan dari malam ke-15 sampai malam ke-21 (15-21). Selanjutnya untuk total malam yang lebih dari 3 minggu (>21)

```python
# Pengelompokkan data berdasarkan hasil pengamatan distribusi data
df2['total_nights_group'] = np.where(df2['total_nights']>21, 21,
                                     np.where(df2['total_nights']>14,15,df2['total_nights']))

# Membuat tabel aggregat
df2_agg = df2.groupby(['hotel', 'is_canceled', 'total_nights_group'])['arrival_date_month'].count().reset_index(name='num_bookings')
df2_pivot = df2_agg.pivot_table(index=['hotel', 'total_nights_group'], columns='is_canceled', values='num_bookings').reset_index()
df2_pivot['%_canceled'] = round(df2_pivot[1]*100.0/(df2_pivot[0]+df2_pivot[1]), 2)

df2_pivot
```

Tujuan dari plot yang ingin ditampilkan adalah untuk membandingkan 2 tipe hotel terhadap durasi menginapnya pelanggan, sehingga digunakan diagram batang untuk melihat tingkat pembatalan pemesanan hotel. Selain itu, kita akan melihat bagaimana durasi menginap memengaruhi tingkat pembatalan pemesanan hotel, sehingga tren untuk data tersebut ikut ditampilkan pula.

```python
plt.figure(figsize=(15,10))
sns.regplot(x='total_nights_group', y='%_canceled', data=df2_pivot[df2_pivot['hotel'] == 'City Hotel'], scatter=False, label='Trend City Hotel')
sns.regplot(x='total_nights_group', y='%_canceled', data=df2_pivot[df2_pivot['hotel'] == 'Resort Hotel'], scatter=False, label='Trend Resort Hotel')
sns.barplot(x='total_nights_group', y='%_canceled', hue='hotel', data=df2_pivot, alpha=0.75)

plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid()
plt.legend(title='Hotel Type', title_fontsize=15, prop={'size':13})

plt.xticks(range(0,17,1), labels=[str(x) for x in range(0,15,1)]+["15-21",">21"])
plt.xlabel('Total Night Booked', fontsize=15)
plt.ylabel('Percentage of Booking Canceled (%)', fontsize=15)
plt.ylim(0,100)
plt.xlim(0.5,16.5)

plt.text(x=0.5, y=110, s="Positive Trend on Total Nights and Percentage of Booking Canceled",
         fontsize=20, fontweight='bold')
plt.text(x=0.5, y=102, s="The longer the total nights booked, the higher the percentage of booking being canceled, \n also City Hotel has more steeper trend compared to Resort Hotel",
         fontsize=18)

plt.tight_layout()
plt.savefig('investigate_hotel_business_v02.png', dpi=200)
```

# 4. Analisis Pengaruh Jarak Waktu Pemesanan terhadap Tingkat Pembatalan Pemesanan Hotel

```python
df3 = df_prep.copy()

sns.histplot(data=df2, x="lead_time", bins=50, color="pink")
```

Berdasarkan pada histogram di atas, dapat dilihat bahwa persebaran data terlihat right-skewed. Oleh karena itu dilakukan pengelompokkan untuk membuat persebarannya menjadi lebih merata. Jarak waktu pemesanan yang mulanya dalam satuan harian, dikelompokkan menjadi satuan bulanan yakni per 30 hari.

```python
# Kategori kolom lead_time

# lead_time dengan value yg cukup besar (lebih dari 400) dikategorisasikan menjadi satu grup
df3['lead_time'] = np.where(df3['lead_time']>400, 401, df3['lead_time'])
bins = list(np.arange(0,max(df3['lead_time'])+30, 30))
labels = ["{0}-{1}".format(str(bins[x]), str(bins[x+1])) for x in range(len(bins[:-1]))]

# groupping lead_time untuk setiap kelipatan 30
df3['lead_time_group'] = pd.cut(df3['lead_time'],bins=bins, include_lowest=True, labels=labels)
```

```python
# Aggregasi data untuk menghitung jumlah pemesanan berdasarkan group lead_time yang sudah dilakukan proses group dan berdasarkan tipe hotel
df3_book = df3.groupby(['hotel','lead_time_group', 'is_canceled'], observed=False)['hotel'].count().reset_index(name='num_bookings')
df3_pivot = df3_book.pivot_table(index=['hotel','lead_time_group'], columns='is_canceled', values='num_bookings', observed=False).reset_index()
df3_pivot.columns = ['hotel_type','lead_time_group', 'not_canceled', 'canceled']
df3_pivot['all_bookings'] = df3_pivot['not_canceled']+df3_pivot['canceled']
df3_pivot['%_canceled'] = df3_pivot['canceled']*100.0/(df3_pivot['all_bookings']) 

df3_pivot.head()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='lead_time_group', y='%_canceled', hue='hotel_type', data=df3_pivot)

plt.tick_params(axis='both', which='major', labelsize=12)
plt.grid()
plt.legend(title='Hotel Type', title_fontsize=14, prop={'size':12}, loc=9)

plt.xlabel('Lead Time Group (by 30 days)', fontsize=14)
plt.ylabel('Percentage of Booking Canceled (%)', fontsize=14)
plt.xlim(-0.5,13.5)
plt.ylim(0,100)

# Highlight lead time 1 bulan
plt.axvline(0.5, ls='--', color='red')
plt.text(x=-0.2, y=80, s='1 Month', fontsize=18, color='red', rotation=90)
plt.stackplot(np.arange(-0.5,1.5,1), [[100]], color='indianred', alpha=0.3)

# Highlight lead time diatas 1 tahun
plt.axvline(8.5, ls='--', color='red')
plt.text(x=10, y=90, s='Around 1 Year', fontsize=18, color='red')
plt.stackplot(np.arange(8.5,14.5,1), [[100]], color='indianred', alpha=0.3)

# Judul dan Keterangan tambahan atas judul
plt.text(x=-0.5, y=108, s="Both Hotels have the Lowest Canceled Ratio within 1 Month of Leads Time",
         fontsize=20, fontweight='bold')
plt.text(x=-0.5, y=102, s="Resort Hotel has quite stagnant (in around 40%) canceled ratio, while City Hotel has quite high canceled \nratio (Above 60%) with leads time around 1 year.",
         fontsize=16)

plt.tight_layout()
plt.savefig('investigate_hotel_business_v03.png', dpi=200)
```
